# California Houses price predection

## Framing the Problem
The goal of this report is to analyze the relationship between various socioeconomic factors and median house values in a given dataset. We aim to understand how specific independent variables, such as median income, population, and number of households, influence the dependent variable, which is the median house value.

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model

housing = pd.read_csv("housing.csv")
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
housing.describe()
```

**Proccesing data:to remove noise and outliers**

```python
housing_preprocessed = housing[housing.median_house_value != 500001]
```

**Creating Seaborn scatter plot of median house value vs. median income**

```python
sns.scatterplot(x="median_income", y="median_house_value",data=housing_preprocessed, color="green")  
plt.title("plot No 1_ Median House Value vs. median_income")
```

Create a linear regression object,,

running linear regressions between median house value vs. median income

```python
model=linear_model.LinearRegression()
model.fit(X=housing_preprocessed[["median_income"]], y=housing_preprocessed[["median_house_value"]])
```

print the intercept and slope, the Mean Squared Error-the loss (MSE) and Mean Absolute Error(MAE) between median house value and median income


```python
# print coefficients and Intercept
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
# print the MSE and MAE on the train set
print(f"Mean squared error: {mean_squared_error(housing_preprocessed[['median_house_value']], model.predict(housing_preprocessed[['median_income']]))}")
print(f"Mean absolute error: {mean_absolute_error(housing_preprocessed[['median_house_value']], model.predict(housing_preprocessed[['median_income']]))}")
```

```python
# plot the regression line against the training data
import matplotlib.pyplot as plt

sns.scatterplot(x="median_income",y="median_house_value", data=housing_preprocessed, color="green")
plt.plot(housing_preprocessed[["median_income"]], model.predict(housing_preprocessed[["median_income"]]), color="yellow",linewidth=4)
plt.title("plot No 1.1_ the regression line against the training data")
plt.show()
```

## median house value vs. population:
**Creating Seaborn scatter plot of median house value vs. population**

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt

sns.scatterplot(x="population", y="median_house_value",data=housing_preprocessed, color="blue")
plt.title("plot No 2_ Median House Value vs. population")
```

running linear regressions between median house value vs. popluation

```python
model1=linear_model.LinearRegression()              #making this model as variable and gave it a name so will not be confused with other models in this page
model1.fit(X=housing[["population"]], y=housing[["median_house_value"]])
```

print the intercept and slope, the Mean Squared Error-the loss (MSE) and Mean Absolute Error(MAE) between median house value and popluation


```python
# print coefficients and Intercept
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(f"Intercept: {model1.intercept_}")
print(f"Coefficients: {model1.coef_}")
# print the MSE and MAE on the train set
print(f"Mean squared error: {mean_squared_error(housing_preprocessed[['median_house_value']], model1.predict(housing_preprocessed[['population']]))}")
print(f"Mean absolute error: {mean_absolute_error(housing_preprocessed[['median_house_value']], model1.predict(housing_preprocessed[['population']]))}")
```

Plotting the resulting line on top of the scatter data median house value and population

```python
# plot the regression line against the training data
import matplotlib.pyplot as plt


sns.scatterplot(x="population",y="median_house_value", data=housing_preprocessed, color="blue" )
plt.plot(housing_preprocessed[["population"]], model1.predict(housing_preprocessed[["population"]]),color="red", linewidth=4)
plt.title("plot No 2.1_ the regression line against the training data")
plt.show()
```

## median house value vs. number of households
**Creating Seaborn scatter plot of median house value vs. number of households**

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="households", y="median_house_value",data=housing_preprocessed, color="orange")
plt.title("plot No 3_ Median House Value vs. households")
```


running linear regressions between median house value vs. number of householders

```python
model2=linear_model.LinearRegression()
model2.fit(X=housing[["households"]], y=housing[["median_house_value"]])
```

print the intercept and slope, the Mean Squared Error-the loss (MSE) and Mean Absolute Error(MAE) between median house value and number of house hold


```python
# print coefficients and Intercept
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(f"Intercept: {model2.intercept_}")
print(f"Coefficients: {model2.coef_}")
# print the MSE and MAE on the train set
print(f"Mean squared error: {mean_squared_error(housing_preprocessed[['median_house_value']], model2.predict(housing_preprocessed[['households']]))}")
print(f"Mean absolute error: {mean_absolute_error(housing_preprocessed[['median_house_value']], model2.predict(housing_preprocessed[['households']]))}")
```

Plotting the resulting line on top of the scatter data median house value and number of households

```python
# plot the regression line against the training data
import matplotlib.pyplot as plt


sns.scatterplot(x="households",y="median_house_value", data=housing_preprocessed, color="orange" )
plt.plot(housing_preprocessed[["households"]], model2.predict(housing_preprocessed[["households"]]),color="red", linewidth=4)
plt.title("plot No 3.1_ the regression line against the training data")
plt.show()
```
